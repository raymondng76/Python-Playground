```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 1000 points between 0, 20
x = np.linspace(0, 20, 1000)
```

```python
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('pyplot')
plt.show()
```

```python
x = np.random.randn(100, 2)
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
x = np.random.randn(200, 2)
x[:50] += 3
y = np.zeros(200)
y[:50] = 1
plt.scatter(x[:,0],x[:,1], c=y)
plt.show()
```
